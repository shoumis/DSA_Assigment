```python
#5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.
```

```python
#Bar Chart
plt.bar(data2['Classification'],data2['SW'])
plt.title('CLASSIFICATION Vs SW')
plt.xlabel('CLASSIFICATION')
plt.ylabel('SW')
plt.show()
```

```python
#Box Plot
plt.boxplot(data2[['SL','SW','PL','PW']])
plt.title('Box Plot')
plt.show()
```

```python
#Histogram
plt.figure(figsize=(8,6))
plt.hist(data2['SL'],color='red',rwidth=0.8,cumulative=True)
plt.title('Distribution of SL',fontsize=16)
plt.xlabel('SL')
plt.ylabel('Frequency')
plt.show()
```

```python
#plotting two subplots in a single plot

plt.figure(figsize=(16,5))

plt.subplot(1,2,1)
plt.hist(data2['SL'],color='blue',rwidth=0.8,cumulative=True)
plt.title('Distribution of SL',fontsize=16)
plt.xlabel('SL')
plt.ylabel('Frequency')

plt.subplot(1,2,2)
plt.hist(data2['SW'],color='green',rwidth=0.8,cumulative=True)
plt.title('Distribution of SW',fontsize=16)
plt.xlabel('SW')
plt.ylabel('Frequency')

plt.tight_layout()
```

```python
#Scatter Plot
plt.figure(figsize=(6,4))
plt.scatter(data2['SL'],data2['PL'],s=20,c='green',marker='^')
plt.title('Scatter Plot of SL vs PL')
plt.xlabel('SL')
plt.ylabel('PL')
plt.show()
```

```python
#plotting two subplots in a single plot
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.scatter(data2['SL'],data2['PL'],s=20,c='green',marker='^')
plt.title('Scatter Plot of SL vs PL')
plt.xlabel('SL')
plt.ylabel('PL')

plt.subplot(1,2,2)
plt.scatter(data2['SW'],data2['PW'],s=20,c='blue',marker='*')
plt.title('Scatter Plot of SW vs PW')
plt.xlabel('SW')
plt.ylabel('PW')

plt.tight_layout()
```

```python
#visualization using panda
data2['SL'].sort_values().plot.bar()
```

```python
data2[['SL','SW']].plot.bar()
```

```python
#Pair Plot
sns.pairplot(data2)
plt.show()
```

```python
#Heat Map
sns.heatmap(data2.corr(),annot=True)
plt.title('Heat Map')
plt.show()
```

```python
#Boxplot using Seaborn
sns.boxplot(x='Classification',y='PW',data=data2)
plt.title('Boxplot of Classification vs PW')
plt.show()
```

```python
#Strip Plot
sns.stripplot(data2['Classification'],data2['SL'])
plt.title('Strip Plot of Classification vs SL')
plt.show()
```
